#Eclipse Install

We will be configuring eclipse to use Java 12 so we can use more switch functionality.  The process will involve three steps:

* Downloading the Java 12 JDK & JRE
* Downloading Eclipse for Java
* Configuring Eclipse

I've also created a simple 'how to' at the end for setting up your first project and class.  I link to an external document about how to link Eclipse with GitHub.




##First, download Java 12:

Go to https://www.oracle.com/java/technologies/javase/jdk12-archive-downloads.html and select the Java 12.0.2 download that matches your platform (Windows, Linux, Mac).  Click to download and except the terms and conditions.

Before your package starts downloading, Oracle will ask you to make an account.  Go ahead and do so.  You can use your @seattlecolleges.edu email.

* Once it is downloaded, double click on the package to run the install script on your system.

##Set up your local Java environment

For Windows or Mac, open a command prompt, then do the following depending on which system you are using.

###Windows


Type the following command to update the `JAVA_HOME` environment variable (requires administrator privilege):

```
setx -m JAVA_HOME "C:\Program Files\Java\jdk-12.0.2"
```
Also make sure to include JAVA_HOME\bin in the PATH variable:
```
setx -m PATH "%PATH%;%JAVA_HOME%\bin";
```
Finally, open a new command prompt and type `java –version` to verify.  It will return your Java install version, which should be 12.0.2.

###Mac: 

Open a command prompt and type:
```
export JAVA_HOME="/Library/Java/JavaVirtualMachines/jdk-12.0.2.jdk/Contents/Home"
```

Then check to make sure the version is 12:

`java -version`

If that worked, then we're ready to make the change perminent.  Figure out if you are running `bash` or `zsh` in your shell by typing
`echo $SHELL`.  If you are running `zsh` the ouptut of this command will be `/bin/zsh` if you are running `bash` the output of this command will be `/bin/bash`.  Follow the appropriate directions below:

###bash

If you're running `bash` open your bash profile at `~/.bash_profile`  and add 
```
export JAVA_HOME="/Library/Java/JavaVirtualMachines/jdk-12.0.2.jdk/Contents/Home"
```
at the bottom of the file without changing anything else.

###zshell
If you’re running `zshell` (`zsh`) open `~/.zshenv` and add
```
export JAVA_HOME="/Library/Java/JavaVirtualMachines/jdk-12.0.2.jdk/Contents/Home"
```
at the bottom without changing anything else.


###Test the bash or zsh change

Open a new terminal window and make sure `java -version` is still says 12 when you run it in the new terminal window.

##Install Eclipse

* Install Eclipse for java developers:   https://www.eclipse.org/downloads/packages/ and select the best download for your system (Mac, Windows, Linux).  
* Install Eclipse by double clicking on the package once it's downloaded.
* Launch Eclipse after it is fully installed.  It will ask you where you would like your working directory to be.  The default is fine.


##Configure Eclipse to use Java 12

Once Eclipse is launched follow the directions for Windows or Mac below.  They are very similar.

###Windows
* Go to Window > Preferences, then navigate to Java > Installed JREs and add the Java 12 JRE (that you installed in Step 1). Tick the box to make it the default JRE.  Here is a link to a doc that contains Windows instructions with pictures.  Skip the install steps and scroll down to the configuration steps.  https://www.zkoss.org/wiki/Setting_Default_JRE_In_Eclipse

###Mac

* Select eclipse > preferences > Installed JRE's and then select the add button.  The location of your JRE will likely be the same as mine, which is `/Library/Java/JavaVirtualMachines/jdk-12.0.2.jdk/Contents/Home/`.  If there's nothing there, go to `/Library/Java/JavaVirtualMachines/` and see which Java versions are available.  Click into Java 12, and navigate to `Home`, that's the location Eclipse will need.  
* A diologue box will pop up and ask you about the JRE type, select Standard VM.
* For JRE home, put `/Library/Java/JavaVirtualMachines/jdk-12.0.2.jdk/Contents/Home/` or the location of your install, if yours is not the same as mine.
* For Name, I reccomend java-12.0.2, though you can choose whatever name you like.
* Hit “Finish”, and you’ll see a new entry under the JRE paths with the name you chose, select java-12.0.2 (or your alternate name), then hit “apply and close”.

### Confirm JRE for Mac and Windows

To confirm it worked go to Run> Run Configuration > JRE and you should see something like: ‘Java SE 12’ (java-12.0.2).  If you chose a different JRE name, that name will be in parenthesis.  

Congrats!  You're ready to go!

##Simple Eclipse Setup
If you've never used Eclipse, here's a quick 'how to' guide to get you started.

* You can safely confirm the default workspace when it asks you.  There's a box you can check to make it stop asking you.

1. To start, make a project: 
* Click on File -> New -> Java Project.  It will 
Name your project whatever you like.  I'm going to go with `Playground`.

* It will automatically pop up a Module dialogue for you. You do not need to create a module at this time.  Skip the module creation
* You'll see your project pop up in the left hand navigation window.

2. Make a class:
* Right click on the `Playground` project and click New > Class.  
* Type the name of your new class.  Let’s call our class Input, so we can test it with our Input.java script.  It will open a coding window with a pre-declared Input class and some minimal code.
* Note, you can configure Eclipse to pre-generate a main method for you too, if you like.

3. Put your old Input.java code from the first two weeks into the Input.java class.  You can go ahead and delete the pre-generated code first.  Here it is if you need it:

```
import java.util.Scanner;

// Goal: Calulator

public class Input {                          
  public static void main(String [] args) {  
    Scanner whatEver = new Scanner(System.in);             // Create a Scanner object

    System.out.println("What is your first number?");             
    int firstNumber = whatEver.nextInt();                  // Get int from user and put in firstNumber

    System.out.println("What is your second number?");
    int secondNumber = whatEver.nextInt();                 // Get second int from user and put in secondNumber
    
    System.out.println("Now we will add your numbers!");
    int myAnswer = firstNumber + secondNumber;             // Add both numbers together and put in myAnswer
                             
   
    System.out.println("My first number is " + firstNumber + ".\nMy second Number is " + secondNumber + ".\nAdded together they equal " + myAnswer + ".");    // Print "My name is" followed by the name the user entered.
  }
}
```
3. Save your code: File > Save

4. Run your code.  Right click “Input.java” in Package Explorer then -> Run As -> Java Application.

You will see a console pop up and it will prompt you for input.  You’re done!  You can create all your new classes under this project for now.  Later in the quarter we will use Eclipse to generate packages and work with Gradle.
